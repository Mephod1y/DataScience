```python
authors = df.groupby(['author']).agg({'name': ['count']})
authors
```

```python
authors.shape
```

```python
max = authors['name'].max()
authors[authors['name','count']==max[0]]
```

```python
authors1 = df.groupby(['author']).agg({'user_rating': ['mean']})
authors1
```

```python
min = authors1['user_rating'].min()
authors1[authors1['user_rating','mean']==min[0]]
```

Який у цього автора середній рейтинг? 3.9

```python
pd.concat([authors, authors1], axis=1).sort_values([('name','count'),('user_rating','mean')])
```
